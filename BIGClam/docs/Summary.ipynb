# Промежуточный итог (09.06)

## Итак, Что у нас есть:

** 4 модели на основе BigClam для выделения пересекающихся сообществ. **
* BigClam
* BigClam с делением на вес (обобщение на взвешенный случай)
* Gamma Model в чистом виде
* Sparse Gamma Model с Гамма моделью на ненулевых весах

*Основные выводы:*
* BigCLAM не так уж просто перебить
* Gamma Model в чистом виде очень долго работает

** Эксперименты на простых модельных данных и 4х эго-графах **
* Проверена адекватность реализаций предложенных методов
* Предложен метод подбора количества сообществ, который на практике работает для небольших графов (очень рано останавливается)
* Предложена регуляризация для декорреляции матрицы F (проблема сильных пересечений в матрице F). Если пользоваться аккуратно -- приносит пользу.
* Небольшой ресеч на тему зависимости качества оптимизации от параметра alpha в методе подбора шага backtrackingLineSearch. Можно динамически переключать, но пока решили не трогать.

*Основные выводы:*
* На простых модельных данных все достаточно точно работает
* Не факт, что гамма распределение адекватно отражает распределение весов в реальных данных, выбрано для удобства

** Эксперименты по инициализации **
* Придумана концепция разреженной инициализации на основе locally minimal neighborhoods и развита до 5 различных вариаций.
* Продемонстрировано, что инициализация в BigClam не самая хорошая (нули в матрице F так нулями могут и остаться, новые методы изначально стартуют из лучшего значения правдоподобия)
* Новая инициализация немного ускоряет и уточняет работу схемы оптимизации

** Эксперименты на реальных данных (зачаток) **
* Сравнение по модулярности на двух небольших взвешенных графах. BigClam с простым шагом победил
* Есть проблемы с выбором шага — backtrackingLineSearch сходится быстрее, но уменьшающийся шаг приводит к лучшему значению. Для Гамма модели это оказывается критичным и backtrackingLineSearch в ней не работает.

** Эксперименты с данными AMiner (aminer.org/billboard/AMinerNetwork) **

*Цель:*

Создать модель по генерированию реальных взвешенных данных на основе графа цитирования с известным разбиением на сообщества.

*Итог:*
* Куча красивых картиночек.
* Данные содержат множество пропусков и неточные названия, графы, которые получаются не содержат нормальной структуры сообществ
* Очень много мелких сообществ, граф получается не связным и прочие проблемы. Возможно, при более детальном изучении данных, этого можно избежать.


## Чего у нас нет, наши проблемы:

* Почти полное отсутствие нормальных данных для тестирования взвешенных методов, особенно с истинным разбиением на сообщества.
* На существующих данных без истинной разметки Новые методы не сильно лучше или хуже, чем BigClam.
* Без разметки можем сравниваться по Модулярности (специфичная штука) и Проводимости (не факт, что хорошо отражается самим BigClam-ом)


## Что можно предпринять

** 1) Перейти в плоскость кластеризации **

* Разделение смеси гауссиан -- модельные данные.
* Взять реальные данные с ядерной метрикой преобразовать в взвешенные данные для Гамма модели, погонять эксперименты посравниваться.
* Попробовать на примерах из scikit learn

** 2) Продолжить разбираться с данными AMiner **

* Построить граф ключевых слов, экспериментировать на нем.
* Докопаться таки до реальной структуры

** 3) Взвесить невзвешенные графы **
* Например, в качестве веса ребра между вершинами брать количество общих вершин в их 1-окрестности, либо другую подобную характеристику

** !4) Разобраться с бенчмарком от Lancinetti, тестироваться на нем **
* Возможно, это хороший вариант в наших реалиях
* Минус: Все основания, что наш метод будет плохо работать, в этом плане плоскость кластеризации может быть предпочтительнее
